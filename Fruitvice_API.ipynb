```python
import requests
import json
import pandas as pd
```

```python
#obtain the fruitvice API data using requests.get("url") function. 
#The data is in a json format.
data = requests.get("https://fruityvice.com/api/fruit/all")
```

```python
#retrieve results using json.loads() function.
results = json.loads(data.text)
```

```python
#convert the json data into pandas data frame.
pd.DataFrame(results)
```

```python
#The result is in a nested json format. 
#The 'nutrition' column contains multiple subcolumns, so the data needs to be 'flattened' or normalized.
df2 = pd.json_normalize(results)
df2.head()
```

```python
#Example: we need to know the family and genus of a apple:
apple = df2.loc[df2["name"] == 'Apple']
(apple.iloc[0]["family"]) , (apple.iloc[0]["genus"])

```

```python
#How many calories are contained in a apple:
apple = df2.loc[df2["name"] == "Apple"]
apple.iloc[0]['nutritions.calories']
```
